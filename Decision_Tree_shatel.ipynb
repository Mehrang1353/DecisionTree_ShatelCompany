```python
import pandas as pd
import numpy as np
```

Import data

```python
data=pd.read_excel('churn2.xlsx')
data
```

Remove some Feature 

```python
data=data.drop('code',axis=1)
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

Chang Type of some feature

```python
li=['city','churn','namava','shoteltalk','casper', 'Android Security']
data[li]=data[li].astype(str)
```

```python
data.dtypes
```

```python
data.describe()
```

Normalization

```python
sp=data['speed']
sp_n=(sp-sp.min(axis=0))/(sp.max(axis=0)-sp.min(axis=0))
data['speed']=sp_n
data
```

validation


```python
from sklearn.model_selection import train_test_split
x=data.drop('churn',axis=1)
y=data['churn']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=101)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

learn with Model decision Tree


```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=10,criterion='gini',min_samples_split=4,min_samples_leaf=3,random_state=101)
dt.fit(x_train,y_train)
dt
```

predict applay Model


```python
y_pred=dt.predict(x_test)
y_pred

```

Evaluation


```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from IPython.display import display
from sklearn.metrics import classification_report
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
con_mat=pd.crosstab(y_test,y_pred,rownames=['ActualData'],colnames=['PredictedData'],margins=True)
display(con_mat)
```

```python
print(classification_report(y_test,y_pred))
```
